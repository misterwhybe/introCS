```python
import math
import random
import scipy.integrate
import numpy
import matplotlib.pyplot as plt
from scipy import signal

### input parameters ####################
# def SIR_model(y,T,Beta,Gamma, N, mu, v):
#     X, Y, Z = y
# #     xi = random.uniform(0.0,1.0)
#     dX_dt = v * N  - (Beta * X * Y /N ) - mu * X
#     dY_dt = (Beta * X * Y /N) - Gamma * Y - mu * Y
#     dZ_dt = Gamma * Y - mu * Z
    
#     return[dX_dt, dY_dt, dZ_dt]

# int; total population
N = 350
X0 = 349
Y0 = N - X0
Z0 = 0
# + random.uniform(0.0,1.0)
# float; maximum elapsed time
# T = numpy.linspace(0,200,10000)
t = 0.0
T = 200.0
# float; spatial parameter
v = 1/60
mu = 1/60

# float; rate of infection after contact
Beta = 10

# float; rate of cure
Gamma = 0.5

birth = mu*N
transmission = Beta*X0*Y0/N
recovery = Gamma * Y0
deathX = mu*X0
deathY = mu*Y0
deathZ = mu*Z0

total = birth + transmission + recovery + deathX + deathY + deathZ
# initialize results list
SIR_data = []
SIR_data.append([t, X0, Y0, Z0])
# main loop
while t < T:

    if Y0 == 0:
           break

    w1 = Beta * X0 * Y0 / N
    w2 = Gamma * Y0
    W = w1 + w2

    dt = -math.log(random.uniform(0.0,1.0)) / W
    t = t + dt
    
    if random.uniform(0.0,1.0) < w1 / W:
        X0 = X0 - 1
        Y0 = Y0 + 1

    else:
        Y0 = Y0 - 1
        Z0 = Z0 + 1
         
    SIR_data.append([t, X0, Y0, Z0])
#########################################
# print(SIR_data[:,1])
# print(SIR_data[:,2])
# print(SIR_data[:,3])
print(SIR_data)
SIR_data = ''.join(SIR_data)
print(SIR_data)

# plt.plot(SIR_data[0], SIR_data[1], label="X(t)")
# plt.plot(SIR_data[0], SIR_data[2], label="Y(t)")
# plt.plot(SIR_data[0], SIR_data[3], label="Z(t)")
    
plt.plot(SIR_data, label="X(t)")
# plt.plot(SIR_data[2], label="Y(t)")
# plt.plot(SIR_data[3], label="Z(t)")

plt.grid()

plt.legend()
plt.xlabel("Time (weeks)")
plt.ylabel("Fraction of population in each class")
plt.title("SIR model")
plt.xlim(0,100)
plt.ylim(0,500)
# place the beta and gamma at the correct place
# plt.text(170, 300, r'Beta = %.2f' %Beta, fontsize=8)
# plt.text(170, 280, r'Gamma = %.2f' %Gamma, fontsize=8)

plt.show()
```
